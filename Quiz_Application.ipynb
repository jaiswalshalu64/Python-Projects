# **Quiz Application - Mini Project**

At the end of the quiz, the total score, correct answers, and incorrect answers are displayed.

**Features:**

Multiple-choice questions with four options.

Tracks total score and provides immediate feedback.

Simple, interactive console-based interface.

**How to Use:**

Run the script in Google Colab or any Python IDE.

Answer questions as prompted.

View your final results at the end.
